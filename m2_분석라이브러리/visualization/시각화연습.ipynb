#### 선그래프

```python
sns.lineplot(data = df, x = "model year", y = "mpg", err_style='bars')
```

- sns.lineplot() 함수에서는 err_style 매개변수를 이용해 밴드의 스타일을 조정할 수 있습니다. 
- 예를 들어, err_style='band'로 설정하면, 투명한 밴드가 그려지고, err_style='bars'로 설정하면, 오차 막대(error bar)가 그려집니다.

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df, x = "model year", y = "mpg", hue = "origin",\
             err_style='band', palette=['red', 'green', 'blue'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.arange(0, 10, 0.1)
y = np.sin(x)

# 선 그래프 그리기
plt.plot(x, y)

# 밴드 그리기
y_upper = y + 0.2  # 밴드 상단 값
y_lower = y - 0.2  # 밴드 하단 값
plt.fill_between(x, y_upper, y_lower, alpha=0.2)

# 그래프 스타일 설정
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Title')
plt.grid(True)

# 그래프 보여주기
plt.show()
```

#### 파이 그래프

```python
df['Ratio'] = 1
dfo = df.groupby('origin').sum()
dfo.Ratio.plot(kind='pie',
                     figsize=(8,6),
                     autopct='%.2f%%',
                     startangle=0)
plt.title('origin', size=20)
plt.axis('equal') # 가로와 세로의 비율을 동일하게 설정
plt.legend(labels=dfo.index,loc='best')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
g = sns.FacetGrid(iris, col='species')
g.map(sns.scatterplot, 'petal_length', 'sepal_length')

# 그래프를 표시합니다.
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# 첫 번째 서브플롯에 산점도를 그립니다.
sns.scatterplot(ax=axes[0, 0], data=iris, x='sepal_length', y='sepal_width', hue='species')
axes[0, 0].set_title('Scatterplot')

# 두 번째 서브플롯에 히스토그램을 그립니다.
sns.histplot(ax=axes[0, 1], data=iris, x='sepal_length', kde=True)
axes[0, 1].set_title('Histogram')

# 세 번째 서브플롯에 박스플롯을 그립니다.
sns.boxplot(ax=axes[1, 0], data=iris, x='species', y='petal_length')
axes[1, 0].set_title('Boxplot')

# 네 번째 서브플롯에 바이올린플롯을 그립니다.
sns.violinplot(ax=axes[1, 1], data=iris, x='species', y='petal_width')
axes[1, 1].set_title('Violinplot')

# 그래프를 표시합니다.
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 첫 번째 서브플롯에 히스토그램을 그립니다 (KDE 포함).
sns.histplot(ax=axes[0], data=iris, x='sepal_length', kde=True)

# 두 번째 서브플롯에 히스토그램을 그립니다 (KDE 미포함).
sns.kdeplot(ax=axes[1], data=iris, x='sepal_length')

# 그래프를 표시합니다.
plt.show()
```

#### 결측치 처리

```python
import numpy as np
import pandas as pd
ar = np.arange(100).reshape(10,10)
df = pd.DataFrame(ar, columns=list('abcdefghij'))
df.iloc[2:5,3:5] = np.nan
df.iloc[1:3,1:3] = np.nan
print(df)
df.isnull().sum()
```

```python
df_null = df.copy()
```

```python
# 결측치 있는 행 삭제
df_null.dropna(inplace=True)
df_null.isnull().sum().sum()
```

```python
# subset은 삭제할 행의 열(column)을 지정하는 리스트이며, inplace는 기본값이 False이며, 
# True로 설정하면 데이터프레임이 업데이트
df_null = df.copy()
df_null.dropna(subset=['d'],inplace=True)
print(df_null)
df_null.isnull().sum().sum()
```

```python
df_null.b.fillna('N',inplace=True)
df_null
```

```python
# 결손데이터 처리하기(평균값으로 대체)
df_null = df.copy()
print(df_null)
df_null.c.replace(np.nan, df_null.c.mean(),inplace=True)
df_null.head()
```

```python
import seaborn as sns

# titanic 데이터셋 가져오기
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
titanic_df.isnull().sum()
```

```python
# deck 열의 NaN 개수 계산하기
titanic_df.deck.value_counts(dropna=False)
```

```python
# deck 열의 NaN 개수 계산하기
nan_deck = titanic_df['deck'].value_counts(dropna=False)
# nan_deck = df['deck'].value_counts()
print(nan_deck,'\n')
print(titanic_df['deck'].isnull().sum(axis=0))
```

```python
titanic_df.info()
```

```python
# thresh = 500 : NaN 값이 500개 이상인 열을 모두 삭제 - deck 열(891개 중 688개의 NaN 값)
# how = 'any': NaN 값이 하나라도 있으면 삭제
# how = 'all': 모든 데이터가 NaN 값일 경우에만 삭제
df_thresh = titanic_df.dropna(axis=1, thresh=500)  
print(df_thresh.columns)
```

```python
# age Null 값 처리 : drop vs. mean 대체
import warnings
warnings.filterwarnings('ignore')

df1 = titanic_df[['age','survived']]
print(df1.age.isnull().sum())
# age NaN 값 버리기
rdf1 = df1.dropna(subset=['age'], how='any', axis=0)
print(rdf1.age.isnull().sum())
print(df1.age.isnull().sum())
# age를 평균값으로 대체
df1['age'].fillna(df1['age'].mean(), inplace=True)
print(df1.age.isnull().sum())
```

```python
rdf = titanic_df[['embarked','sex']]
rdf.embarked.value_counts()
```

```python
rdf = titanic_df[['embarked','sex']]
most_freq = rdf['embarked'].value_counts(dropna=True).idxmax()
most_freq
```

```python
# embarked 열의 NaN 값을 승선도시 중에서 가장 많이 출현한 값으로 변환
rdf = titanic_df[['embarked','sex']]
most_freq = rdf['embarked'].value_counts(dropna=True).idxmax()
rdf['embarked'].fillna(most_freq, inplace=True)
rdf.info()
```

```python
rdf.embarked.value_counts()
```

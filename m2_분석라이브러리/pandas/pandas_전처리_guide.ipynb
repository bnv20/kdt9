```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head(3)
```

```python
df.age.unique()
```

```python
def get_category(x):
    cat = ''
    if x <= 10:
        cat = 'Child'
    elif x <= 30:
        cat = 'Young Adult'
    elif x <=60:
        cat = 'Adult'
    else:
        'Elderly'
    return cat

df['age_class'] = df.age.apply(lambda x:get_category(x))
df.loc[:,['age','age_class']].head()
```
